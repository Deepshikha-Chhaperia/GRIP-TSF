```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('SampleSuperstore.csv')

# Step 2: Data Cleaning
# Check for missing values
missing_values = df.isnull().sum()

# Check for duplicates
duplicates = df.duplicated().sum()

# Summary statistics
summary_stats = df.describe()

# Step 3: Data Visualization
# Visualize the distribution of Sales and Profit
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')

plt.subplot(1, 2, 2)
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title('Distribution of Profit')
plt.show()

# Visualize profit by category and sub-category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=df, palette='viridis')
plt.title('Profit by Category')
plt.xticks(rotation=45)

plt.figure(figsize=(12, 6))
sns.barplot(x='Sub-Category', y='Profit', data=df, palette='viridis')
plt.title('Profit by Sub-Category')
plt.xticks(rotation=90)
plt.show()

# Step 4: Business Problem Identification
# Analyze profit by region
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print("Profit by Region:\n", profit_by_region)

# Identify low-profit products
low_profit_products = df[df['Profit'] < 0][['Category', 'Sub-Category', 'Sales', 'Profit']]
print("\nLow-Profit Products:\n", low_profit_products)
```
